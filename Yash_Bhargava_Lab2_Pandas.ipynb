**read_csv()** method in pandas is used to load a csv file into pandas dataFrame

---


**to_string()** method is used to print all the complete rows without skipping 
middle rows.

```python
df = pd.read_csv("/content/drive/MyDrive/data_panda.csv")
print(df.head(169).to_string())
```

dataFrame head() method takes x, as it' s parameter, where x are the number of rows to be taken from beginning.

---
By default it takes out 5 rows from the dataset if not parameter is taken.


---
Syntax:- df.head(x)

```python
df1 = df.head(169)
print(df1)
```

dataFrame tail() method takes x, as it' s parameter, where x is the number of rows to be taken from the end of dataset.

---
By default it takes out 5 rows from the dataset if not parameter is taken.


---
Syntax:- df.tail(x)

```python
df2 = df1.tail()
print(df2)
```

dropna() method of dataFrame removes the rows that contain NaN values.

---


syntax:- df.dropna()

---


it has a optional attribute inplac = True, if set to True then makes changes to original dataFrame otherwise returns a new dataFrame.

```python
df3 = df1.dropna()
print(df3)
```

fillna() method of dataFrame replaces the non numeric values with some desired values.

```python
df4 = df.fillna(130)
print(df4)
```

Replacing the non numeric values of 'Calories' column with mean of the column.

```python
df5 = df
df5['Calories'].fillna(df['Calories'].mean(), inplace = True)
```

```python
print(df5)
```

```python
print(df5)
```

Replacing the non numeric values of 'Calories' column with median of the column.

```python
df6 = df5
df6['Calories'].fillna(df['Calories'].median(), inplace = True)
print(df6)
```

Replacing the non numeric values of 'Calories' column with mode of the column.

```python
df7 = df5
df7['Calories'].fillna(df['Calories'].mode()[0], inplace = True)
print(df7)
```

iloc[] is used to fetch the columns on the based of indexes from complete dataFrame.

```python
df8 = df7
print(df8.iloc[[2,5,7]])
```

loc[] is used to fetch columns from complete dataFrame on the basis of column headings.

```python
print((df.loc[:168, 'Duration':'Maxpulse']).to_string())
```

The index property returns the index information of the DataFrame.

```python
for x in df8.index:
  if df8.loc[x,"Duration"] > 60:
    df8.loc[x,"Duration"] = 80
print(df8)
```

Pandas drop_duplicates() method helps in removing duplicates from the Pandas Dataframe In Python.

```python
print(df8.drop_duplicates(inplace = True))
print(df8.to_string())
```
