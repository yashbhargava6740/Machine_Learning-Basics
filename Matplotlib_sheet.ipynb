```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
x1 = np.array([9,10,11,12])
y1 = np.array([13,14,15,16])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x,y, marker = 'o')
plt.show()
```

```python
plt.plot(x,y,marker = 's')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Graph")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.xlabel("Graph-1-X")
plt.ylabel("Graph-1-Y")
plt.subplot(2,1,2)
plt.plot(x1,y1)
plt.xlabel("Graph-2-X")
plt.ylabel("Graph-2-Y")
plt.show()
```

```python
plt.suptitle("Graphs")
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Graph-1")
plt.xlabel("Graph-1-X")
plt.ylabel("Graph-1-Y")
plt.subplot(1,2,2)
plt.plot(x1,y1)
plt.title("Graph-2")
plt.xlabel("Graph-2-X")
plt.ylabel("Graph-2-Y")
plt.show()
```
